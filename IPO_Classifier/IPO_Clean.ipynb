```python
#import libraries

import yfinance as yf
import pandas as pd
```

```python
#Load in Data

raw_df = pd.read_csv('IPO.csv', encoding= 'unicode_escape')
raw_df
```

```python
#Taking only the features we need to conduct analysis

df = raw_df.loc[:, ["Symbol", "Year", "Month", "Day", "dayOfWeek", "closeDay0", "highDay0",
                    "openDay0", "lowDay0", "volumeDay0", "closeDay260", "highDay260", "openDay260",
                    "lowDay260", "volumeDay260", "Sector", "Industry", "exactDateFounded"]]
df
```

```python
#analyzing the data

df.describe()
```

```python
#dropped over 1000 rows, but now we have congruity with 2564 rows with all the data

df = df.dropna()

df.describe()
```

```python
#converting year, month, date columns to a single column named IPODATE
#converting IPODATE into Y - M - D, A format that yfinance can read

df['IPODATE'] = pd.to_datetime(df[['Year', 'Month', 'Day']])
df['IPODATE'] = df['IPODATE'].apply(lambda x: x.strftime('%Y-%m-%d'))
```

```python
#making list from IPODATE for easier conversion

ipoDATE = df['IPODATE'].tolist()
ipoDATE
```

```python
# Generating 6 month return
six_month_return = []

for date in ipoDATE:
    date = pd.to_datetime(date)
    start_date = date - pd.DateOffset(months=6)
    end_date = date
    sp500 = yf.download('^GSPC', start=start_date, end=end_date)
    six_month_return.append((sp500['Adj Close'][-1] - sp500['Adj Close'][0]) / sp500['Adj Close'][0])
    
#REF1
```

```python
# Generating 3 month return
three_month_return = []


for date in ipoDATE:
    date = pd.to_datetime(date)
    start_date = date - pd.DateOffset(months=3)
    end_date = date
    sp500 = yf.download('^GSPC', start=start_date, end=end_date)
    three_month_return.append((sp500['Adj Close'][-1] - sp500['Adj Close'][0]) / sp500['Adj Close'][0])
    
# REF1
```

```python
#creating columns for the market returns of 3 and 6 months

df = df.assign(markReturn3mo=three_month_return)
df = df.assign(markReturn6mo=six_month_return)
```
